```python
# 1) detectar la columna 'region' sin cambiar nombres
def find_region_col(df): #Recorre el nombre de las columnas del DataFrame
    for c in df.columns:
        if str(c).strip().lower() == "region":
            return c
    raise KeyError("No encuentro la columna 'region'.")

#A cada columna lo convierte a string, le saca espacios con el strip() y lo
# pasa a minuscula
# esta función busca cómo se llama exactamente la columna región en la tabla
# recorre todas las columnas utilizando el dor y a cada nombre lo pasa a minusula y
# le saca los espacios; si queda "region" devuelve ese nombre original c.

colR_2005 = find_region_col(base2005)
colR_2025 = find_region_col(base2025)

# esto guarda el nombre real de la columna region en cada base

# 2) máscara para GBA (01)
def mask_gba(df, col):
    s = df[col]
    return (
        pd.to_numeric(s, errors="coerce").eq(1) |
        s.astype(str).str.strip().str.zfill(2).eq("01") |
        s.astype(str).str.strip().str.lower().isin(["gran buenos aires","g.b.a","gba"])
    )
# Esta función devuelve True las filas que son de GBA y False e las que no
# si la región viene con un 1, 01 o con una palabra  la marca como GBA
# usamos el astype("string") para que no se rompa si hay numeros o NaN mezclados.


# 3) filtrar SIN tocar ninguna otra columna
base2005_gba = base2005.loc[mask_gba(base2005, colR_2005)].copy()
base2025_gba = base2025.loc[mask_gba(base2025, colR_2025)].copy()

# El .loc[mask] se queda con las filas True de arriba y hace una copia

# 4) verlas “como el original”, solo filtradas
base2005_gba.head()
base2025_gba.head()

base2005_gba.to_csv("base2005_gba.csv")
base2025_gba.to_csv("base2025_gba.csv")
```

```python
import pandas as pd
import numpy as np
import unicodedata

# === Partimos de la base original (SIN renombres)
df = base_gba1.copy()

# ---------- NORMALIZACIÓN PREVIA DE ch04 (SEXO) ----------
def _strip_accents(s: str) -> str:
    return ''.join(c for c in unicodedata.normalize('NFKD', s) if not unicodedata.combining(c))

def normalizar_sexo(col):
    """
    Devuelve serie numérica: 1=Varón / 2=Mujer, a partir de
    valores 1/2 (numéricos o strings) o descripciones textuales.
    """
    s = col.copy()

    # a) Pasar todo a string, trim, minúscula y sin acentos
    s = s.astype(str).str.strip()
    s = s.replace({"": pd.NA, "nan": pd.NA, "None": pd.NA})
    s_norm = s.dropna().map(lambda x: _strip_accents(str(x)).lower())

    out = pd.Series(pd.NA, index=col.index, dtype="Int64")

    # b) Casos ya codificados como 1/2 (incluyendo strings "1","2")
    mask_1 = s.isin([1, "1"])
    mask_2 = s.isin([2, "2"])
    out.loc[mask_1] = 1
    out.loc[mask_2] = 2

    # c) Textuales
    if not s_norm.empty:
        # varón / hombre / masculino
        mask_varon = s_norm.str.contains(r"\b(varon|hombre|mascul)\b", regex=True)
        # mujer / femenino
        mask_mujer = s_norm.str.contains(r"\b(mujer|femen)\b", regex=True)
        out.loc[mask_varon.index[mask_varon]] = 1
        out.loc[mask_mujer.index[mask_mujer]] = 2

    return out

# Si existe ch04, la normalizamos ANTES de cualquier otra limpieza
if "ch04" in df.columns:
    df["ch04"] = normalizar_sexo(df["ch04"])

# --- Reglas conservadoras (EPH) ---
# 1) INGRESOS: negativos -> NaN
ingresos_vars = [c for c in ["ipcf", "itf"] if c in df.columns]

# 2) EDAD (ch06): fuera de [0, 110] -> NaN
edad_var = "ch06" if "ch06" in df.columns else None

# 3) BINARIAS 1/2: fuera de {1,2} -> NaN
binarias_12 = [c for c in ["ch09", "pp02i", "ch10", "pp06a", "ch08"] if c in df.columns]

# 4) SEXO (ch04): fuera de {1,2} -> NaN  (ya normalizada arriba, acá sólo validamos)
sexo_var = "ch04" if "ch04" in df.columns else None

# 5) OTRAS CATEGÓRICAS: negativos -> NaN (regla mínima segura)
otras_categoricas = [c for c in ["nivel_ed","estado","cat_inac","pp04a","pp04g"] if c in df.columns]

# --- Copia para limpiar y log de cambios
subset_clean = df.copy()
changes = {}

# 1) Ingresos
for v in ingresos_vars:
    subset_clean[v] = pd.to_numeric(subset_clean[v], errors="coerce")
    mask = subset_clean[v] < 0
    changes[f"{v}_negativos"] = int(mask.sum())
    subset_clean.loc[mask, v] = pd.NA

# 2) Edad
if edad_var:
    subset_clean[edad_var] = pd.to_numeric(subset_clean[edad_var], errors="coerce")
    mask = (subset_clean[edad_var] < 0) | (subset_clean[edad_var] > 110)
    changes[f"{edad_var}_fuera_rango"] = int(mask.sum())
    subset_clean.loc[mask, edad_var] = pd.NA

# 3) Binarias 1/2
for v in binarias_12:
    subset_clean[v] = pd.to_numeric(subset_clean[v], errors="coerce")
    mask = subset_clean[v].notna() & ~subset_clean[v].isin([1,2])
    changes[f"{v}_fuera_1_2"] = int(mask.sum())
    subset_clean.loc[mask, v] = pd.NA

# 4) Sexo 1/2 (validación)
if sexo_var:
    # ya está en Int64 por la normalización; sólo validamos
    mask = subset_clean[sexo_var].notna() & ~subset_clean[sexo_var].isin([1,2])
    changes[f"{sexo_var}_fuera_1_2"] = int(mask.sum())
    subset_clean.loc[mask, sexo_var] = pd.NA

# 5) Otras categóricas (negativos)
for v in otras_categoricas:
    subset_clean[v] = pd.to_numeric(subset_clean[v], errors="coerce")
    mask = subset_clean[v] < 0
    changes[f"{v}_negativos"] = int(mask.sum())
    subset_clean.loc[mask, v] = pd.NA

# --- Resultado final
base_gba1_clean = subset_clean #armo una nueva base de datos con las variables limpias

print("Recodificaciones efectuadas (celdas afectadas):")
for k, v in changes.items():
    if v > 0:
        print(f"  - {k}: {v}")
print(f"\nBase limpia -> base_gba1_clean  |  shape: {base_gba1_clean.shape}")

# (chequeo rápido opcional para que veas que NO quedan NaN masivos en 2005)
if "ano4" in base_gba1_clean.columns and "ch04" in base_gba1_clean.columns:
    print("\nDistribución ch04 por año (post-limpieza):")
    print(base_gba1_clean.groupby("ano4")["ch04"].value_counts(dropna=False))

# Guardar
base_gba1_clean.to_csv("base_gba1_clean.csv", index=False)
base_gba1_clean.head(15000)
```

```python
# 1

# === Configuración de base y columnas ===
df = base_gba1_clean1              # tu dataframe principal
edad_col = "ch06"                  # edad
pobre_col = "itf_bajo_cbt_hogar"   # 1 = pobre, 0 = no pobre (ajusta si usaste otro nombre)

# 1) Crear edad^2
df["edad2"] = pd.to_numeric(df[edad_col], errors="coerce")**2

# 2) Panel A: histograma de edad
import numpy as np
import matplotlib.pyplot as plt

edad = pd.to_numeric(df[edad_col], errors="coerce").dropna()

plt.figure()
plt.hist(edad, bins=range(0, int(np.nanmax(edad))+2, 2), edgecolor="black")
plt.title("Panel A — Histograma de edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

# 3) Panel B: KDE de edad (pobres vs no pobres), usando un KDE simple sin dependencias
def kde1d(x, grid=None, bw=None):
    x = np.asarray(x, dtype=float)
    x = x[np.isfinite(x)]
    n = x.size
    if n == 0:
        if grid is None: grid = np.linspace(0, 100, 200)
        return grid, np.zeros_like(grid)
    if grid is None:
        lo = np.nanpercentile(x, 1)
        hi = np.nanpercentile(x, 99)
        if not np.isfinite(lo) or not np.isfinite(hi) or lo >= hi:
            lo, hi = 0.0, max(100.0, float(np.nanmax(x)))
        grid = np.linspace(lo, hi, 300)
    if bw is None:
        std = np.nanstd(x, ddof=1)
        bw = 1.06 * std * n**(-1/5) if std > 0 else 1.0
    inv = 1.0 / (bw * np.sqrt(2*np.pi))
    z = (grid[:, None] - x[None, :]) / bw
    dens = inv * np.exp(-0.5 * z**2)
    dens = dens.mean(axis=1)
    return grid, dens

pobres = pd.to_numeric(df.loc[df[pobre_col]==1, edad_col], errors="coerce").dropna()
nopob  = pd.to_numeric(df.loc[df[pobre_col]==0, edad_col], errors="coerce").dropna()

# mismo grid para comparar
grid_min = np.nanpercentile(edad, 1)
grid_max = np.nanpercentile(edad, 99)
grid = np.linspace(grid_min, grid_max, 300)

g_p, d_p = kde1d(pobres, grid=grid)
g_n, d_n = kde1d(nopob,  grid=grid)

plt.figure()
plt.plot(g_p, d_p, label="Pobres")
plt.plot(g_n, d_n, label="No pobres")
plt.title("Panel B — Densidad (KDE) de edad: pobres vs no pobres")
plt.xlabel("Edad")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```

```python
#5
import pandas as pd
import numpy as np

df = base_gba1_clean1.copy()

# Año en numérico y orden deseado
df["ano4"] = pd.to_numeric(df["ano4"], errors="coerce").round().astype("Int64")
years = [2005, 2025]

# Helpers
def add_total(s):
    s = s.reindex(years, fill_value=0)
    return pd.Series([int(s.get(2005,0)), int(s.get(2025,0)), int(s.sum())],
                     index=[2005, 2025, "Total"])

def safe_pct(num, den):
    num, den = float(num), float(den)
    return np.nan if den == 0 else round(100*num/den, 2)

# Conteos básicos
n_obs      = df.groupby("ano4").size()
na_pobre   = df.groupby("ano4")["itf_bajo_cbt_hogar"].apply(lambda s: s.isna().sum())
n_pobres   = df.groupby("ano4")["itf_bajo_cbt_hogar"].apply(lambda s: (s==1).sum())
n_nopobres = df.groupby("ano4")["itf_bajo_cbt_hogar"].apply(lambda s: (s==0).sum())

# Número de variables limpias/homogeneizadas (columnas de la base final)
n_vars = df.shape[1]
n_vars_row = pd.Series([n_vars, n_vars, n_vars + n_vars], index=[2005, 2025, "Total"])

# === TABLA 1 ===
tabla1 = pd.DataFrame({
    "Cantidad observaciones":                                   add_total(n_obs),
    "Cantidad de observaciones con NAs en la variable 'Pobre'": add_total(na_pobre),
    "Cantidad de Pobres":                                       add_total(n_pobres),
    "Cantidad de No Pobres":                                    add_total(n_nopobres),
    "Cantidad de variables limpias y homogeneizadas":           n_vars_row
}).T.astype("Int64")

print("Tabla 1 — Resumen de la base final (GBA)")
print(tabla1)

# (Opcional) exportar
# tabla1.to_csv("tabla1_resumen_gba.csv")

# === Comentario breve automático ===
N2005, N2025, NTOT = tabla1.loc["Cantidad observaciones"]
NA2005, NA2025, NATOT = tabla1.loc["Cantidad de observaciones con NAs en la variable 'Pobre'"]
P2005, P2025, PTOT = tabla1.loc["Cantidad de Pobres"]
NP2005, NP2025, NPTOT = tabla1.loc["Cantidad de No Pobres"]

print("\nComentario:")
print(f"- Tamaño total: {NTOT} obs. (2005: {N2005}; 2025: {N2025}).")
print(f"- NAs en 'Pobre': {NATOT} ({safe_pct(NATOT, NTOT)}%); "
      f"2005: {NA2005} ({safe_pct(NA2005, N2005)}%), 2025: {NA2025} ({safe_pct(NA2025, N2025)}%).")
print(f"- Pobres: {PTOT} y No pobres: {NPTOT} "
      f"(participación pobres total: {safe_pct(PTOT, PTOT+NPTOT)}%).")
print(f"- Variables limpias/homogeneizadas en la base final: {n_vars}.")
```

```python
#ej 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ====== 0) Tomo tu base de PERSONAS (GBA) ======
df = base_gba1_clean1.copy()   # <-- cambialo si tu df se llama distinto

# ----- helpers -----
def to_num_clean(s: pd.Series) -> pd.Series:
    if s.dtype == "O":
        s = (s.astype(str)
               .str.replace(r"[^\d\-,\.]", "", regex=True)
               .str.replace(".", "", regex=False)
               .str.replace(",", ".", regex=False))
    return pd.to_numeric(s, errors="coerce")

# ----- columnas robustas -----
df.columns = [c.strip() for c in df.columns]
L = {c.lower(): c for c in df.columns}
pick = lambda *n: next(L[k.lower()] for k in n if k.lower() in L)

codusu   = pick("CODUSU")
edad_col = pick("edad","CH06")
educ_col = pick("educ","NIVEL_ED","nivel_ed")
itf_col  = pick("ITF","itf")
pp3e     = pick("PP3E_TOT","pp3e_tot")
pp3f     = pick("PP3F_TOT","pp3f_tot")

# ====== 1) Variables pedidas ======
# IX_TOT desde conteo de personas por CODUSU
df["IX_TOT"] = df.groupby(codusu)[codusu].transform("size")

# horastrab = pp3e_tot + pp3f_tot
df[pp3e] = pd.to_numeric(df[pp3e], errors="coerce")
df[pp3f] = pd.to_numeric(df[pp3f], errors="coerce")
df["horastrab"] = df[pp3e].fillna(0) + df[pp3f].fillna(0)

# edad, edad2, educ, ITF
df["edad"]  = pd.to_numeric(df[edad_col], errors="coerce")
df["edad2"] = df["edad"]**2
df["educ"]  = pd.to_numeric(df[educ_col], errors="coerce")
df["ITF"]   = to_num_clean(df[itf_col])

vars_pca = ["edad","edad2","educ","ITF","IX_TOT","horastrab"]
X = df[vars_pca].dropna().copy()

# ====== 2) Estandarización ======
scaler = StandardScaler()
Z = scaler.fit_transform(X)   # z-scores

# ====== 3) PCA ======
pca = PCA(n_components=2, random_state=0)
scores = pca.fit_transform(Z)              # índices (scores)
expl = pca.explained_variance_ratio_       # varianza explicada

# Loadings (contribución de cada variable a cada componente)
loadings = pd.DataFrame(
    pca.components_.T,
    index=vars_pca,
    columns=["PC1","PC2"]
)

print("Varianza explicada:", {"PC1": expl[0], "PC2": expl[1], "Acumulada": expl.sum()})
print("\nLoadings (pesos de variables en cada PC):")
display(loadings.round(3))

# ====== 4) Scatter de scores PC1 vs PC2 ======
fig, ax = plt.subplots(figsize=(7,5.4), dpi=130, constrained_layout=True)
ax.scatter(scores[:,0], scores[:,1], s=6, alpha=0.6)
ax.set_xlabel(f"PC1 ({expl[0]*100:.1f}% var.)")
ax.set_ylabel(f"PC2 ({expl[1]*100:.1f}% var.)")
ax.set_title("PCA con ingreso – Scores PC1 vs PC2")
ax.axhline(0, linewidth=0.5); ax.axvline(0, linewidth=0.5)
plt.show()

# (opcional) guardar para el informe
pd.DataFrame(scores, columns=["PC1_score","PC2_score"]).to_csv("scores_pca.csv", index=False)
loadings.to_csv("loadings_pca.csv")
print("Archivos guardados: scores_pca.csv, loadings_pca.csv")
```

```python
# EJERICIO 3 . BIPLOT PCA — flechas SIN textos + leyenda arriba con colores
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ===== 0) Base =====
df = base_gba1_clean1.copy()

def to_num_clean(s: pd.Series) -> pd.Series:
    if s.dtype == "O":
        s = (s.astype(str)
               .str.replace(r"[^\d\-,\.]", "", regex=True)
               .str.replace(".", "", regex=False)
               .str.replace(",", ".", regex=False))
    return pd.to_numeric(s, errors="coerce")

# columnas robustas (case-insensitive)
df.columns = [c.strip() for c in df.columns]
L = {c.lower(): c for c in df.columns}
pick = lambda *names: next(L[n.lower()] for n in names if n.lower() in L)

codusu   = pick("CODUSU")
edad_col = pick("edad", "CH06")
educ_col = pick("educ", "NIVEL_ED", "nivel_ed")
itf_col  = pick("ITF", "itf")
pp3e     = pick("PP3E_TOT", "pp3e_tot")
pp3f     = pick("PP3F_TOT", "pp3f_tot")

# ===== 1) Variables =====
df["IX_TOT"] = df.groupby(codusu)[codusu].transform("size")
df[pp3e] = pd.to_numeric(df[pp3e], errors="coerce")
df[pp3f] = pd.to_numeric(df[pp3f], errors="coerce")
df["horastrab"] = df[pp3e].fillna(0) + df[pp3f].fillna(0)

df["edad"]  = pd.to_numeric(df[edad_col], errors="coerce")
df["edad2"] = df["edad"]**2
df["educ"]  = pd.to_numeric(df[educ_col], errors="coerce")
df["ITF"]   = to_num_clean(df[itf_col])

vars_pca = ["edad","edad2","educ","ITF","IX_TOT","horastrab"]
X = df[vars_pca].dropna().copy()

# ===== 2) Standardize + PCA (2 PCs) =====
Z = StandardScaler().fit_transform(X)
pca = PCA(n_components=2, random_state=0).fit(Z)
scores   = pca.transform(Z)
loadings = pca.components_.T
expl     = pca.explained_variance_ratio_

# Escala de flechas
scale_scores = scores.std(axis=0)
arrow = loadings * scale_scores * 2.0

# Colores pedidos y etiquetas para la leyenda de arriba
color_map = {
    "edad":      "red",      # rojo
    "edad2":     "purple",   # violeta
    "educ":      "green",    # verde
    "ITF":       "yellow",   # amarillo
    "IX_TOT":    "orange",   # naranja
    "horastrab": "black",    # negro
}
nice = {
    "edad": "Edad",
    "edad2": "Edad (cuadrado)",
    "educ": "Nivel de educación alcanzado",
    "ITF": "Ingreso total familiar",
    "IX_TOT": "Número de miembros en el hogar",
    "horastrab": "Horas trabajadas",
}

# ===== 3) Biplot (sin textos en flechas) =====
fig, ax = plt.subplots(figsize=(8.4, 7.4), dpi=130)

# puntos
ax.scatter(scores[:,0], scores[:,1], s=8, alpha=0.08, color="#5a7da1")

# ejes guía
ax.axhline(0, lw=0.7, ls="--", color="gray")
ax.axvline(0, lw=0.7, ls="--", color="gray")

# flechas de colores (SIN texto)
for var, (vx, vy) in zip(vars_pca, arrow):
    ax.arrow(0, 0, vx, vy, color=color_map[var],
             lw=2.6, head_width=0.08, length_includes_head=True)

ax.set_xlabel(f"PC1 ({expl[0]*100:.1f}% varianza explicada)")
ax.set_ylabel(f"PC2 ({expl[1]*100:.1f}% varianza explicada)")
ax.set_title("Biplot PCA", pad=22)

# Leyenda arriba con circulitos de color → variable
handles = [Line2D([0],[0], marker="o", linestyle="",
                  markerfacecolor=color_map[v], markeredgecolor=color_map[v],
                  markersize=10, label=nice[v]) for v in vars_pca]
ax.legend(handles=handles, loc="upper center", bbox_to_anchor=(0.5, 1.12),
          ncol=3, frameon=False, title=None)

plt.tight_layout()
plt.show()
```

```python
#4
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# === Usa tu DataFrame final de PERSONAS (cambiá el nombre si hace falta) ===
df = base_gba1_clean1.copy()

# --- helpers ---
def to_num_clean(s):
    if s.dtype == "O":
        s = (s.astype(str)
               .str.replace(r"[^\d\-,\.]", "", regex=True)
               .str.replace(".", "", regex=False)
               .str.replace(",", ".", regex=False))
    return pd.to_numeric(s, errors="coerce")

# --- columnas robustas ---
df.columns = [c.strip() for c in df.columns]
L = {c.lower(): c for c in df.columns}
pick = lambda *n: next(L[k.lower()] for k in n if k.lower() in L)

codusu   = pick("CODUSU")
edad_col = pick("edad","CH06")
educ_col = pick("educ","NIVEL_ED","nivel_ed")
itf_col  = pick("ITF","itf")
pp3e     = pick("PP3E_TOT","pp3e_tot")
pp3f     = pick("PP3F_TOT","pp3f_tot")

# --- variables pedidas ---
df["IX_TOT"] = df.groupby(codusu)[codusu].transform("size")
df[pp3e] = pd.to_numeric(df[pp3e], errors="coerce")
df[pp3f] = pd.to_numeric(df[pp3f], errors="coerce")
df["horastrab"] = df[pp3e].fillna(0) + df[pp3f].fillna(0)

df["edad"]  = pd.to_numeric(df[edad_col], errors="coerce")
df["edad2"] = df["edad"]**2
df["educ"]  = pd.to_numeric(df[educ_col], errors="coerce")
df["ITF"]   = to_num_clean(df[itf_col])

vars_pca = ["edad","edad2","educ","ITF","IX_TOT","horastrab"]
X = df[vars_pca].dropna()

# --- estandarizar + PCA 6 componentes ---
Z = StandardScaler().fit_transform(X)
pca = PCA(n_components=6, random_state=0).fit(Z)
expl = pca.explained_variance_ratio_
cum  = expl.cumsum()

# 1) Scree plot: varianza explicada por PC (EN LÍNEA)
x = range(1, len(expl) + 1)
fig1, ax1 = plt.subplots(figsize=(7.2, 5.0))
ax1.plot(x, expl, marker="o")  # línea con marcadores
ax1.set_xticks(list(x))
ax1.set_ylim(0, max(expl) * 1.15)
ax1.set_xlabel("Componente principal")
ax1.set_ylabel("Proporción de varianza explicada")
ax1.set_title("Scree plot – Varianza explicada por componente (6 PCs)")
for i, v in enumerate(expl, start=1):
    ax1.text(i, v + 0.01, f"{v*100:.1f}%", ha="center", fontsize=8)
plt.show()

# 2) Varianza explicada acumulada
fig2, ax2 = plt.subplots(figsize=(7.2, 5.0))
ax2.plot(range(1,7), cum, marker="o")
ax2.set_xticks(range(1,7))
ax2.set_ylim(0, 1.05)
ax2.set_xlabel("Número de componentes")
ax2.set_ylabel("Proporción acumulada de varianza explicada")
ax2.set_title("Varianza explicada acumulada (6 PCs)")
for i,v in enumerate(cum, start=1):
    ax2.text(i, v+0.02, f"{v*100:.1f}%", ha="center", fontsize=8)
plt.show()

# (opcional) tabla para el informe
pd.DataFrame({"PC":[f"PC{i}" for i in range(1,7)],
              "Varianza_exp":expl,
              "Varianza_acum":cum}).round(4)
```

```python
#5
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from collections import Counter

# ========= 0) Base =========
df = base_gba1_clean1.copy()   # <-- ajustá el nombre si hace falta

# ---- helper para ITF en string ----
def to_num_clean(s: pd.Series) -> pd.Series:
    if s.dtype == "O":
        s = (s.astype(str)
               .str.replace(r"[^\d\-,\.]", "", regex=True)
               .str.replace(".", "", regex=False)
               .str.replace(",", ".", regex=False))
    return pd.to_numeric(s, errors="coerce")

# ---- columnas robustas ----
df.columns = [c.strip() for c in df.columns]
L = {c.lower(): c for c in df.columns}
pick = lambda *n: next(L[k.lower()] for k in n if k.lower() in L)

edad_col = pick("edad","CH06")
itf_col  = pick("ITF","itf")
pobre_col = "itf_bajo_cbt_hogar"  # tu etiqueta (1=pobre, 0=no)

# ========= 1) Variables y estandarización =========
df["edad"] = pd.to_numeric(df[edad_col], errors="coerce")
df["ITF"]  = to_num_clean(df[itf_col])

Xraw = df[["edad","ITF"]].dropna().copy()

scaler = StandardScaler()
Z = scaler.fit_transform(Xraw)     # z-scores para KMeans

# ========= 2) KMeans y gráficos para k=2,4,10 =========
def plot_clusters(Z, Xraw, k, ax):
    km = KMeans(n_clusters=k, n_init=20, random_state=0)
    lab = km.fit_predict(Z)
    # centros en escala original para graficar
    centers = scaler.inverse_transform(km.cluster_centers_)
    ax.scatter(Xraw["edad"], Xraw["ITF"], c=lab, s=6, alpha=0.6)
    ax.scatter(centers[:,0], centers[:,1], s=120, marker="X")  # centroides
    sil = silhouette_score(Z, lab)
    ax.set_title(f"k-means (k={k}) – silhouette={sil:.2f}")
    ax.set_xlabel("Edad"); ax.set_ylabel("Ingreso total familiar (ITF)")
    return lab, km, sil

fig, axes = plt.subplots(1, 3, figsize=(15, 4.8), constrained_layout=True)
labels_k2, km2, sil2 = plot_clusters(Z, Xraw, 2,  axes[0])
labels_k4, km4, sil4 = plot_clusters(Z, Xraw, 4,  axes[1])
labels_k10,km10,sil10= plot_clusters(Z, Xraw, 10, axes[2])
plt.show()

# ========= 3) ¿k=2 separa pobres vs no pobres? =========
# Alinear la verdad con el mismo subset usado para cluster (mismo índice)
y = df.loc[Xraw.index, pobre_col]

# Normalizar a {1=pobre,0=no} por si viniera raro
if y.dtype == "O":
    y = (y.astype(str).str.lower()
           .str.normalize("NFKD").str.encode("ascii","ignore").str.decode("ascii"))
    y = y.replace({"si":"1","sí":"1","no":"0","nopobre":"0"})
    y = y.map(lambda s: 1 if "1" in s or "pobre" in s else (0 if "0" in s or "no" in s else np.nan))
else:
    y = y.map({1:1, 0:0, 2:0})

mask = y.notna()
y_true = y[mask].astype(int)
labs2  = pd.Series(labels_k2, index=Xraw.index)[mask]

# Mapear cada cluster a la clase mayoritaria (0/1)
map_cluster = {c: Counter(y_true[labs2==c]).most_common(1)[0][0] for c in np.unique(labs2)}
y_pred = labs2.map(map_cluster)

acc = (y_pred.values == y_true.values).mean()
cm  = pd.crosstab(y_true, y_pred, rownames=["Real (1=pobre)"], colnames=["Predicho (k=2)"])
print(f"Exactitud k=2 vs '{pobre_col}': {acc:.3f}")
display(cm)
```

```python
# === Dendrograma jerárquico con etiquetas grandes (tamaños de hoja) ===
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram

df = base_gba1_clean1.copy()

# -- helper para limpiar numéricos --
def to_num_clean(s: pd.Series) -> pd.Series:
    if s.dtype == "O":
        s = (s.astype(str)
               .str.replace(r"[^\d\-,\.]", "", regex=True)
               .str.replace(".", "", regex=False)
               .str.replace(",", ".", regex=False))
    return pd.to_numeric(s, errors="coerce")

# -- resolver nombres reales de columnas (case-insensitive) --
df.columns = [c.strip() for c in df.columns]
L = {c.lower(): c for c in df.columns}
pick = lambda *alts: next(L[a.lower()] for a in alts if a.lower() in L)

# Edad
edad_col = pick("edad","ch06")
df["edad"] = pd.to_numeric(df[edad_col], errors="coerce")
df["edad2"] = df["edad"]**2

# Educación
educ_col = pick("educ","nivel_ed")
df["educ"] = pd.to_numeric(df[educ_col], errors="coerce")

# ITF
itf_col = pick("ITF","itf")
df["ITF"] = to_num_clean(df[itf_col])

# IX_TOT (miembros del hogar)
cod_col = pick("codusu")  # si tu columna se llama distinto, agregalo como alternativa
df["IX_TOT"] = df.groupby(cod_col)[cod_col].transform("size")

# horastrab = pp3e_tot + pp3f_tot
pp3e = pick("pp3e_tot")
pp3f = pick("pp3f_tot")
df["horastrab"] = pd.to_numeric(df[pp3e], errors="coerce").fillna(0) + \
                  pd.to_numeric(df[pp3f], errors="coerce").fillna(0)

# --- matriz numérica con TODAS las filas (imputamos NaN con mediana) ---
vars_use = ["edad","edad2","educ","ITF","IX_TOT","horastrab"]
num = df[vars_use].apply(pd.to_numeric, errors="coerce")
X_full = pd.DataFrame(SimpleImputer(strategy="median").fit_transform(num),
                      columns=vars_use, index=num.index)
N_eff = len(X_full)

# Elegí el tamaño promedio deseado por hoja -> números más grandes
tam_prom_hoja = 800      # ↑ subilo para ver números aún más grandes (p.ej. 1000)
p = max(5, int(round(N_eff / tam_prom_hoja)))

# --- linkage y dendrograma truncado, ordenado por tamaño ---
Z = StandardScaler().fit_transform(X_full)
link = linkage(Z, method="ward")

plt.figure(figsize=(12, 6))
d = dendrogram(
    link,
    truncate_mode="lastp",
    p=p,
    show_leaf_counts=True,
    count_sort="descending",     # hojas de mayor a menor tamaño
    leaf_rotation=60,
    leaf_font_size=11,
    color_threshold=0.7 * link[:, 2].max()
)

# Dejar SOLO el número entre paréntesis como etiqueta
labels_only = [lab[lab.find("("):] if "(" in lab else lab for lab in d["ivl"]]
ax = plt.gca()
ax.set_xticklabels(labels_only, rotation=60, ha="right")
ax.set_title("Dendrograma - Clustering jerárquico (hojas ordenadas por tamaño)")
ax.set_xlabel("Clusters hoja (tamaño entre paréntesis)")
ax.set_ylabel("Distancia")
plt.tight_layout()
plt.show()

print(f"N efectivo usado = {N_eff} | p = {p} | tamaño promedio por hoja ≈ {N_eff//p}")
```
